```python
def resumen_conectividad(row):
    servicios_internet = ['ADSL', 'CABLEMODEM', 'FIBRAOPTICA', '4G', '3G', 'WIRELESS', "DIALUP", "TELEFONIAFIJA", "SATELITAL"]
    for servicio in servicios_internet:
        if not pd.isna(row[servicio]):
            return "SI"
    return "NO"
```
